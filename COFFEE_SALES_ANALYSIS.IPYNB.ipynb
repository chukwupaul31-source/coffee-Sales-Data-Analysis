```python
import pandas as pd
```

```python
df = pd . read_csv ('coffee_sales_data_location2.csv')
```

```python
df.head ()
```

```python
df.columns = df.columns. str.strip()
df
```

```python
df.rename (columns={
    'Coffee Type':'Coffee_Type', 'Payment Type':'Payment_Type'
},inplace = True)
df
```

```python
df['Time'] = pd. to_datetime (df['Time'],errors='coerce')
df
```

```python
df.reset_index (drop= True, inplace =True)
df
```

```python
# converting price to numeric
```

```python
df['Price'] = pd.to_numeric (df['Price'],errors = 'coerce')
df
```

```python
# checking for null values
```

```python
print(df.isnull().sum())
```

```python
# converting date and time to date time
```

```python
df ['Date'] = pd.to_datetime (df['Date'])
df
```

```python
df ['Time'] =pd.to_datetime  (df['Time'], format ='%h:%m:%s')
df
```

```python
# creating a new column
```

```python
df['Day'] = df ['Date']. dt.day_name()
df
```

```python
# getting mondat, Tuesday
```

```python
df['Month']  =  df['Date'].dt.month_name()
df
```

```python
# january, and february
```

```python
df['Hour'] = pd.to_datetime (df['Time'], format = '%h:%m:%s').dt.hour
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x=sales_by_type.index, y=sales_by_type.values)
plt.title("Total Sales by Coffee Type")
plt.ylabel("Total Sales ($)")
plt.xlabel("Coffee Type")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x=sales_by_day.index, y=sales_by_day.values)
plt.title("Total Sales by Day of the Week")
plt.ylabel("Total Sales ($)")
plt.xlabel("Day")
plt.show()
```

```python
sales_by_hour = df.groupby('Hour')['Price'].sum()

plt.figure(figsize=(8,5))
sns.lineplot(x=sales_by_hour.index, y=sales_by_hour.values, marker='o')
plt.title("Hourly Sales Trend")
plt.xlabel("Hour of the Day")
plt.ylabel("Total Sales ($)")
plt.show()
```

```python
sales_by_type = df.groupby('Coffee_Type')['Price'].sum().sort_values(ascending=False)

plt.figure(figsize=(8,5))
sns.barplot(x=sales_by_type.index, y=sales_by_type.values)
plt.title("Total Sales by Coffee Type")
plt.xlabel("Coffee_Type")
plt.ylabel("Total Sales ($)")
plt.xticks(rotation=45)
plt.show()
```

```python
payment_share = df.groupby('Payment_Type')['Price'].sum().sort_values(ascending=False)

plt.figure(figsize=(6,6))
plt.pie(payment_share, labels=payment_share.index, autopct='%1.1f%%')
plt.title("Payment Method Share")
plt.show()
```

```python
sales_by_month = df.groupby('Month')['Price'].sum()

plt.figure(figsize=(8,5))
sns.barplot(x=sales_by_month.index, y=sales_by_month.values)
plt.title("Total Sales by Month")
plt.xlabel("Month")
plt.ylabel("Total Sales ($)")
plt.show()
```

```python
sales_by_day = df.groupby('Day')['Price'].sum()

plt.figure(figsize=(8,5))
sns.barplot(x=sales_by_day.index, y=sales_by_day.values)
plt.title("Sales by Day of the Week")
plt.xlabel("Day")
plt.ylabel("Total Sales ($)")
plt.show()
```
